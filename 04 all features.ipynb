# Data Pre-processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the dataset
df = pd.read_csv("data.csv")
df
```

```python
#check missing values
print("\nMising Value:\n",df.isnull().sum())

#check datatype for each columns
print("\nDatatype for each column\n",df.dtypes)

#check duplicate values
df.duplicated().sum()
```

```python
sns.countplot(x="diagnosis", data=df) # checking whether value realistic or not
plt.show()
```

```python
# dropping duplicate values
df.drop_duplicates(inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

cleaned_df = df.copy()

# Convert categorical variables to numerical format using label encoding
categorical_cols = ['diagnosis'] #tukar column diagnosis

#check missing values
print("\nMising Value:\n",cleaned_df.isnull().sum())

# dropping id column
cleaned_df.drop('id',axis=1, inplace=True)
cleaned_df.drop('Unnamed: 32',axis=1, inplace=True)

#check datatype for each columns
print("\nDatatype for each column\n",cleaned_df.dtypes) #pastikan semua numerical data

#tanya dr noureen nak bahagi 3 atau lumpsum sekali
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier


# Define features and target variable
feature_df = cleaned_df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

x = np.asarray(feature_df)
y = np.asarray(cleaned_df['diagnosis']) # Target

# Split dataset into 80% training and 20% testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)  # Stratify to balance classes

# Scale features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42)
x_train_balanced, y_train_balanced = smote.fit_resample(x_train, y_train) # type: ignore
```

# Model Development

## SVM

```python
# Create an SVM model with
clf = SVC(kernel='linear', C=1)

# Train the model
SvmModel = clf.fit(x_train_scaled, y_train)

# Test the model
y_pred = clf.predict(x_test_scaled)

# Evaluate model performance
svm_accuracy = accuracy_score(y_test, y_pred)
svm_precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
svm_recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
svm_f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

print("SVM Model Performance:")
print("Accuracy:", svm_accuracy)
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1-score:", svm_f1)
```

## XGBoost

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Encode label (huruf -> nombor)
le = LabelEncoder()
y = le.fit_transform(y)

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# Scaling (optional tapi disarankan)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Create an XGBoost model
clf = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False
)

# Train the model
XgbModel = clf.fit(x_train_scaled, y_train)

# Test the model
y2_pred = clf.predict(x_test_scaled)

# Evaluate model performance
xgb_accuracy = accuracy_score(y_test, y2_pred)
xgb_precision = precision_score(y_test, y2_pred, average='weighted', zero_division=1)
xgb_recall = recall_score(y_test, y2_pred, average='weighted', zero_division=1)
xgb_f1 = f1_score(y_test, y2_pred, average='weighted', zero_division=1)

print("XGBoost Model Performance:")
print("Accuracy:", xgb_accuracy)
print("Precision:", xgb_precision)
print("Recall:", xgb_recall)
print("F1-score:", xgb_f1)
```

## Hybrid using VotingClassifier

```python
# Import libraries
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# --- Encode label (jika belum numeric) ---
le = LabelEncoder()
y = le.fit_transform(y)  # 'B'/'M' -> 0/1

# --- Split data ---
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

# --- Scaling (SVM perlu scaling, XGBoost tak perlu, tapi kita boleh scale dua-dua) ---
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# --- Base Models ---
svm_clf = SVC(kernel='rbf', C=1, probability=True, random_state=42)
xgb_clf = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False
)

# --- Hybrid Model (Voting Classifier) ---
# Soft voting = guna probabiliti (lebih stabil untuk classification)
hybrid_model = VotingClassifier(
    estimators=[('svm', svm_clf), ('xgb', xgb_clf)],
    voting='soft'
)

# --- Train Hybrid Model ---
HybridModel = hybrid_model.fit(x_train_scaled, y_train)

# --- Test Model ---
y3_pred = HybridModel.predict(x_test_scaled)

# --- Evaluate Model Performance ---
hybrid_accuracy = accuracy_score(y_test, y3_pred)
hybrid_precision = precision_score(y_test, y3_pred, average='weighted', zero_division=1)
hybrid_recall = recall_score(y_test, y3_pred, average='weighted', zero_division=1)
hybrid_f1 = f1_score(y_test, y3_pred, average='weighted', zero_division=1)

print("Hybrid (SVM + XGBoost) Model Performance:")
print("Accuracy :", hybrid_accuracy)
print("Precision:", hybrid_precision)
print("Recall   :", hybrid_recall)
print("F1-score :", hybrid_f1)
```
